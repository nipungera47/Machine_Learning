```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0,5,0.1)
y = x ** 3
y2 = x ** 2


# plt.plot(x,y,color = "blue", marker = "p")
plt.plot(x,y,color = "blue",linewidth = 5,label = 'x^3')
plt.plot(x,y2,color = "red",linewidth = 5,label = 'x^2')
plt.ylabel("X^3")
plt.xlabel("x")
plt.title("Demo")
plt.legend()
plt.axis([0,10,0,200])
plt.grid()
# Add Random Text
plt.text(2,80,"text")
plt.show()
```
